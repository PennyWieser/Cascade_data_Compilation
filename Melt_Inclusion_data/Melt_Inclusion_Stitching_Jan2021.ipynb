```python
# This finds file names for all csvs from a path (here MIs_Stitched)
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = os.listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]

mscifile = find_csv_filenames('Data_Jan2020') # Reads in all file names from this directory
print(mscifile)

print('Beginning database creation...')
```

```python
# This takes csvs using file names and makes df. fp is the list of names in mscifile
dfs = {splitext(basename(fp))[0].split('_')[0] : pd.read_csv('Data_Jan2020/{}'.format(fp), 
       error_bad_lines = False, encoding = 'Latin-1') for fp in mscifile}
print (dfs)

print('Database compiled!')
```

```python
csvs = list(dfs.keys())
    
for fp in csvs:
    print(fp)
    print(dfs[fp].columns)
    print(len(dfs[fp].columns))
```

```python
dfs
```

```python
cols = []

print(cols)

#Next, going to clean up column names. Strip all white space, make all lowercase,
#replace spaces, parantheses, backslahes, and percentage signs. Will make 
#column-level data manoiuplation easier in matplotlib/seaborn

print('Beginning column formatting...')

for fp in csvs: 
    dfs[fp].columns = dfs[fp].columns.str.strip().str.lower().str.replace(' ', '_').\
        str.replace('(', '').str.replace(')', '').str.\
        replace('\t', '').str.replace('/', '_').str.replace('%','').\
        str.replace('.', '').str.replace('?','').str.replace('___', '_').\
        str.replace('__', '_')
    lst = list(dfs[fp].columns)    
    cols.append(lst)
print(cols)
print('Column name formatting successful')

```

```python
# Doing this for reference dataframe too
df0.columns=df0.columns.str.strip().str.lower().str.replace(' ', '_').\
        str.replace('(', '').str.replace(')', '').str.\
        replace('\t', '').str.replace('/', '_').str.replace('%','').\
        str.replace('.', '').str.replace('?','').str.replace('___', '_').\
        str.replace('__', '_')
```

```python
for fp in csvs:
#    dfs[fp] = dfs[fp].reindex_like(df0) - This only makes the same number of rows you give it
    dfs[fp] = dfs[fp].reindex(df0.columns, axis=1)
```

```python
dfm = pd.concat(dfs, axis = 0, sort = False)

print('Creating first merged DataFrame')
  
dfm = pd.concat(dfs, axis = 0, sort = False)

print('Creation successful!')

print("Exporting first database - before filtering")

dfm.to_csv('MeltInclusions_Jan2021_Compiled.csv')

print('Database exported!')
```
