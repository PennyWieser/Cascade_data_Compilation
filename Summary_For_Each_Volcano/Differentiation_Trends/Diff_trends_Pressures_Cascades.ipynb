```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
from pickle import load
import pickle
```

 ## Loading Cpx data


```python
with open(r'../Combo_Cpx.pkl', 'rb') as f:
            All_Cpx=load(f)
All_Cpx['MgnoCpx']=(All_Cpx['MgO_Cpx']/40.4044)/((All_Cpx['MgO_Cpx']/40.4044)+(All_Cpx['FeOt_Cpx']/71.844))
```

```python
import numpy as np

uniq_volc = All_Cpx['Volcano'].unique()

# split volcanoes into groups of 3
volcano_groups = np.array_split(uniq_volc, np.ceil(len(uniq_volc) / 3))

color=['red', 'blue', 'cyan', 'grey', 'black', 'yellow']
shape=['s', 'o', 'p', '>', '+','o']

# iterate over each group of volcanoes
for volcano_group_index, volcano_group in enumerate(volcano_groups):

    # create a figure for this group
    fig = plt.figure(figsize=(10,12))

    # iterate over each volcano in the group
    for i, volc in enumerate(volcano_group):
        ax1 = fig.add_subplot(3, 2, 2*i+1)  # Odd indices for ax1
        ax2 = fig.add_subplot(3, 2, 2*i+2, sharey=ax1)  # Even indices for ax2, sharing y with ax1

        ax1.xaxis.set_tick_params(which='both', labelbottom=True)
        ax2.xaxis.set_tick_params(which='both', labelbottom=True)
        fig_save_name='Cpxtrends_Group'+str(volcano_group_index+1)

        Combo_Cpx=All_Cpx.loc[All_Cpx['Volcano']==volc]

        j = 0
        for ref in Combo_Cpx['Reference'].unique():
            study=Combo_Cpx['Reference']==ref
            if volc=='South Sister':
                ax1.plot(Combo_Cpx['MgnoCpx'], Combo_Cpx['P_Wang21_eq1'], shape[j], mfc=color[j], mec='k', label='Waters et al. 2021')
                ax2.plot(Combo_Cpx['MgnoCpx'], Combo_Cpx['P_Jorgenson22'], shape[j], mfc=color[j], mec='k', label='Waters et al. 2021')
            
            else:
                ax1.plot(Combo_Cpx['MgnoCpx'].loc[study], Combo_Cpx['P_Wang21_eq1'].loc[study], shape[j], mfc=color[j], mec='k', label=ref)
                ax2.plot(Combo_Cpx['MgnoCpx'].loc[study], Combo_Cpx['P_Jorgenson22'].loc[study], shape[j], mfc=color[j], mec='k', label=ref)

            j += 1

        ax1.legend()
        ax1.set_title(volc)  # individual title for each subplot
        ax1.set_xlabel('Mg# Cpx')
        ax1.set_ylabel('Pressure (kbar) Wang et al. (2021)')
        ax2.set_title(volc)
        ax2.set_xlabel('Mg# Cpx')
        ax2.set_ylabel('Pressure (kbar) Jorgenson et al. (2022)')
        ax1.set_xlim([0.64, 0.92])

    fig.tight_layout()
    fig.subplots_adjust(top=0.95)  # make some room for the suptitle
    fig.savefig(fig_save_name, dpi=200)
```
