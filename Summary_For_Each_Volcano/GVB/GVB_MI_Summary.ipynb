```python
import pandas as pd
import numpy as np
import os
from os.path import splitext, basename
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reads in the compiled dataset
dfm=pd.read_excel('MeltInclusionCompilation_april23.xlsx')
```

```python
rho=2700
P_axis_min=-3
P_axis_max=12
D_axis_min=100000*P_axis_min/(9.8*rho)
D_axis_max=100000*P_axis_max/(9.8*rho)
```

```python
GarivaldiLake=dfm.loc[dfm['volcano']=="GarivaldiLake"]
GarivaldiLake['Sat_P_Pref']=GarivaldiLake['satp_meash2omeasco2']/1000
GarivaldiLake['Sat_P_Raman']=GarivaldiLake['satp_meash2oramanco2']/1000

GarivaldiLake['Study']="Garibaldi Lake"
GarivaldiLake['Comp']="Basalt"

MountCayley=dfm.loc[dfm['volcano']=="MountCayley"]
MountCayley['Sat_P_Pref']=MountCayley['satp_meash2omeasco2']/1000
MountCayley['Sat_P_Raman']=MountCayley['satp_meash2oramanco2']/1000
MountCayley['Study']="Mount Cayley"
MountCayley['Comp']="Basalt"

MountMeager=dfm.loc[dfm['volcano']=="MountMeager"]
MountMeager['Sat_P_Pref']=MountMeager['satp_meash2omeasco2']/1000
MountMeager['Sat_P_Raman']=MountMeager['satp_meash2oramanco2']/1000
MountMeager['Study']="Mount Meager"
MountMeager['Comp']="Basalt"

SalalGlacier=dfm.loc[dfm['volcano']=="SalalGlacier"]
SalalGlacier['Sat_P_Pref']=SalalGlacier['satp_meash2omeasco2']/1000
SalalGlacier['Sat_P_Raman']=SalalGlacier['satp_meash2oramanco2']/1000

SalalGlacier['Study']="Salal Glacier"
SalalGlacier['Comp']="Basalt"

BridgeRiver=dfm.loc[dfm['volcano']=="BridgeRiver"]
BridgeRiver['Sat_P_Pref']=BridgeRiver['satp_meash2omeasco2']/1000
BridgeRiver['Sat_P_Raman']=BridgeRiver['satp_meash2oramanco2']/1000
BridgeRiver['Study']="Bridge River"
BridgeRiver['Comp']="Basalt"

MI_Combo_allH2O=pd.concat([GarivaldiLake, MountCayley,MountMeager, SalalGlacier, BridgeRiver ], axis=0)
MI_Combo=MI_Combo_allH2O
```

```python
len(MI_Combo)
```

```python
import seaborn as sns
fig, (ax1) = plt.subplots(1, 1, figsize=(5,4), sharey=True)
ax2 = ax1.twinx()

volc_order=["Garibaldi Lake", "Mount Cayley", "Mount Meager", "Salal Glacier", "Bridge River" ]
violin_parts=sns.violinplot(x="Study", y="Sat_P_Pref",  cut=0, inner="stick", linewidth=1,
              hue='Comp',
            data=MI_Combo,  ax=ax1,  order=volc_order, stick=True)
##

volc_order=["Garibaldi Lake", "Mount Cayley", "Mount Meager", "Salal Glacier", "Bridge River" ]
violin_parts=sns.violinplot(x="Study", y="Sat_P_Raman",  cut=0, inner="stick", linewidth=1,
              color='red', facealpha=0.5,
            data=MI_Combo,  ax=ax1,  order=volc_order, stick=True)

ax1.plot([-1, 5], [0, 0], '-k', lw=0.5)
ax1.set_ylim([P_axis_max, P_axis_min])
ax2.set_ylim([D_axis_max, D_axis_min])

ax1.tick_params(direction="in")
ax2.tick_params(direction="in")
ax1.set_xticklabels([])
ax1.set_xlabel(None)
ax1.set_ylabel(None)
ax1.set_xlim([-0.5, 4.5])
#ax1.set_xticklabels(['GarivaldiLake', 'W2016', 'W2019', ''], rotation=90)
ax1.grid(color = 'k', linestyle = '--', axis='y',linewidth = 1, alpha = 0.1, zorder=0)
ax1.get_legend().remove()   
#ax2.set_yticks([])
fig.savefig('GVB_MIs_Glassonly.png', dpi=200, bbox_inches="tight")
```

## What about a fancy half and half plot

```python
MI_Depths_NoRaman=pd.DataFrame(data={'Volcano': MI_Combo['Study'],
                             'Sat_P': MI_Combo['Sat_P_Pref'],
                            'Phase': "NoRaman"})

MI_Depths_Raman=pd.DataFrame(data={'Volcano': MI_Combo['Study'],
                             'Sat_P': MI_Combo['Sat_P_Raman'],
                            'Phase': "Raman"})

MI_Depths=pd.concat([MI_Depths_NoRaman, MI_Depths_Raman], axis=0)
```

```python
import seaborn as sns
fig, (ax1) = plt.subplots(1, 1, figsize=(5,4), sharey=True)
ax2 = ax1.twinx()

volc_order=["Garibaldi Lake", "Mount Cayley", "Mount Meager", "Salal Glacier", "Bridge River" ]
violin_parts=sns.violinplot(x="Volcano", y="Sat_P",  cut=0, inner="stick", linewidth=1,
              hue='Phase', split=True, palette=['royalblue', 'cyan'],
            data=MI_Depths,  ax=ax1,  order=volc_order, stick=True)
##



ax1.plot([-1, 5], [0, 0], '-k', lw=0.5)
ax1.set_ylim([P_axis_max, P_axis_min])
ax2.set_ylim([D_axis_max, D_axis_min])

ax1.tick_params(direction="in")
ax2.tick_params(direction="in")
ax1.set_xticklabels([])
ax1.set_xlabel(None)
ax1.set_ylabel(None)
ax1.set_xlim([-0.5, 4.5])
#ax1.set_xticklabels(['GarivaldiLake', 'W2016', 'W2019', ''], rotation=90)
ax1.grid(color = 'k', linestyle = '--', axis='y',linewidth = 1, alpha = 0.1, zorder=0)
ax1.get_legend().remove()   
#ax2.set_yticks([])
fig.savefig('GVB_MIs_Glassonly.png', dpi=200, bbox_inches="tight")
```
