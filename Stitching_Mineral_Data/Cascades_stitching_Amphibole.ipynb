```python
import Thermobar as pt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Mt Baker

```python
import warnings as w
with w.catch_warnings():
    w.simplefilter('ignore')
    Baker_read_Amp=pt.import_excel('Baker.xlsx', sheet_name='Amp_All')
    Baker_Input_Amp=Baker_read_Amp['my_input']
    Baker_Amp=Baker_read_Amp['Amps']
    Baker_Amp_copy=Baker_Amp.copy()

    Baker_Amp_comps=pt.get_amp_sites_from_input(amp_comps=Baker_Amp)
    Baker_Amp_filt1=Baker_Amp_comps['cation_sum_All'].between(15, 16)
    Baker_Amp_class=pt.classify_phases(df=Baker_Amp, return_end_members=True,
                            str_to_drop='_Amp')

    print(Baker_Amp_class['Phase_Min_Group_ML'].unique())

    Baker_Amp_filt=Baker_Amp.loc[(Baker_Amp_class['Phase_Min_Group_ML']=="Amp")
                                & (Baker_Amp_filt1)].reset_index(drop=True)
    Baker_Input_Amp_filt=Baker_Input_Amp.loc[(Baker_Amp_class['Phase_Min_Group_ML']=="Amp")
                                & (Baker_Amp_filt1)]
    Baker_Amp_px=Baker_Amp.loc[Baker_Amp_class['Phase_Min_Group_ML']=="Px"]

    Baker_Amp_Press_All=pt.calculate_amp_only_press_all_eqs(amp_comps=Baker_Amp_filt,
                        plot=True)
    
    Baker_Input_Amp_filt.to_clipboard(excel=True)
    print(Baker_Input_Amp_filt['Reference'].unique())
```
